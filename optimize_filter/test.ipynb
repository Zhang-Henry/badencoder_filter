```python
import numpy as np
import cupy as cp
import ot

import torch

def euclidean_cost_matrix_torch(embeddings1, embeddings2):
    # 扩展 embeddings1 和 embeddings2 以便计算两两组合的距离
    # embeddings1.shape = [n, features], embeddings2.shape = [m, features]
    # diff.shape = [n, m, features]
    diff = embeddings1.unsqueeze(1) - embeddings2.unsqueeze(0)

    # 计算欧几里得距离，结果 shape = [n, m]
    cost_matrix = torch.sqrt(torch.sum(diff ** 2, dim=-1))

    return cost_matrix


# 示例数据（以 NumPy 数组的形式）
embeddings1 = torch.rand(8,100)
embeddings2 = torch.rand(8,100)  # 第二个分布的直方图

# 假设 embeddings1 和 embeddings2 是两组图像嵌入
cost_matrix = euclidean_cost_matrix_torch(embeddings1, embeddings2)


# 将数据转换为 CuPy 数组以在 GPU 上运行
hist1_gpu = cp.asarray(embeddings1)
hist2_gpu = cp.asarray(embeddings2)
cost_matrix_np = cost_matrix.cpu().detach().numpy()
cost_matrix_gpu = cp.asarray(cost_matrix_np)

# 使用 POT 库计算 Wasserstein 距离
# 注意：部分 POT 函数可能不完全支持 GPU 计算，需要检查特定函数的文档
distance = ot.emd2(hist1_gpu, hist2_gpu,cost_matrix_gpu)

# 将结果转回 CPU
distance_cpu = cp.asnumpy(distance)
print(distance_cpu)
```

```python
import torch

# import vision transformer

from vit_pytorch.simple_vit_with_patch_dropout import SimpleViT
from vit_pytorch.extractor import Extractor

vit = SimpleViT(
    image_size = 256,
    patch_size = 32,
    num_classes = 1000,
    dim = 1024,
    depth = 6,
    heads = 16,
    mlp_dim = 2048,
    patch_dropout = 0.5  # https://arxiv.org/abs/2212.00794
)

vit = Extractor(vit, return_embeddings_only = True, detach = False)

# extractor will enable it so the vision transformer returns its embeddings

# import CoCa and instantiate it

from coca_pytorch.coca_pytorch import CoCa

coca = CoCa(
    dim = 512,                     # model dimension
    img_encoder = vit,             # vision transformer - image encoder, returning image embeddings as (batch, seq, dim)
    image_dim = 1024,              # image embedding dimension, if not the same as model dimensions
    num_tokens = 20000,            # number of text tokens
    unimodal_depth = 6,            # depth of the unimodal transformer
    multimodal_depth = 6,          # depth of the multimodal transformer
    dim_head = 64,                 # dimension per attention head
    heads = 8,                     # number of attention heads
    caption_loss_weight = 1.,      # weight on the autoregressive caption loss
    contrastive_loss_weight = 1.,  # weight on the contrastive loss between image and text CLS embeddings
).cuda()

# mock text and images

text = torch.randint(0, 20000, (4, 512)).cuda()
images = torch.randn(4, 3, 256, 256).cuda()

# train by giving CoCa your text and images with `return_loss = True`

loss = coca(
    text = text,
    images = images,
    return_loss = True  # set this to True to get the full caption + contrastive loss
)

loss.backward()

# do the above for as much text and images...
# then you can get the caption logits as so

logits = coca(
    text = text,
    images = images
) # (4, 512, 20000)

# and the CLIP-like text and image embeddings as

text_embeds, image_embeds = coca(
    text = text,
    images = images,
    return_embeddings = True
) # (4, 512), (4, 512)
```

```python
import torchvision.models as models
import torch.nn as nn
import torch.nn.functional as F
import torch

resnet = models.resnet50(pretrained=True)

class ResNetFeatureExtractor(nn.Module):
    def __init__(self, resnet, selected_layers):
        super(ResNetFeatureExtractor, self).__init__()
        self.selected_layers = selected_layers
        self.pre_layers = nn.Sequential(*list(resnet.children())[:selected_layers[0]])
        self.layers = list(resnet.children())[selected_layers[0]:selected_layers[-1] + 1]

    def forward(self, x):
        x = self.pre_layers(x)
        outputs = []
        for i, layer in enumerate(self.layers):
            x = layer(x)
            if i + self.selected_layers[0] in self.selected_layers:
                outputs.append(x)
        return outputs


# 创建一个特征提取器实例，例如提取第2、3、4个残差块的输出
extractor = ResNetFeatureExtractor(resnet, selected_layers=[4, 5, 6])

def gram_matrix(input):
    a, b, c, d = input.size()
    features = input.view(a * b, c * d)
    G = torch.mm(features, features.t())
    return G.div(a * b * c * d)

def style_loss(generated_features, style_features, style_weights):
    loss = 0
    for gen_feat, style_feat, weight in zip(generated_features, style_features, style_weights):
        G_generated = gram_matrix(gen_feat)
        G_style = gram_matrix(style_feat)
        layer_loss = F.mse_loss(G_generated, G_style)
        loss += layer_loss * weight
    return loss

input_image=torch.rand(4,3,64,64)
features = extractor(input_image)
gram_matrices = [gram_matrix(feature) for feature in features]
```

```python
import torch
import torchvision.models as models
from utils import SinkhornDistance
import torch.nn.functional as F
WD=SinkhornDistance(eps=0.1, max_iter=100)

class ResNetFeatureExtractor(torch.nn.Module):
    def __init__(self, resnet):
        super(ResNetFeatureExtractor, self).__init__()
        # 将ResNet的层分开
        self.conv1 = resnet.conv1
        self.bn1 = resnet.bn1
        self.relu = resnet.relu
        self.maxpool = resnet.maxpool

        self.layer1 = resnet.layer1
        self.layer2 = resnet.layer2
        self.layer3 = resnet.layer3
        self.layer4 = resnet.layer4

    def forward(self, x):
        # 提取各个层的特征
        x = self.conv1(x)
        x = self.bn1(x)
        x = self.relu(x)
        conv1_out = x

        x = self.maxpool(x)
        x = self.layer1(x)
        layer1_out = x

        x = self.layer2(x)
        layer2_out = x

        x = self.layer3(x)
        layer3_out = x

        x = self.layer4(x)
        layer4_out = x

        # 返回提取的特征
        return conv1_out, layer1_out, layer2_out, layer3_out, layer4_out

def gram_matrix(input):
    a, b, c, d = input.size()  # batch size(=1), feature map number, dimensions
    features = input.view(a * b, c * d)  # resize F_XL into \hat F_XL
    G = torch.mm(features, features.t())  # compute the gram product

    # normalize the values of the gram matrix by dividing by the number of element in each feature maps.
    return G.div(a * b * c * d)


def compute_style_loss(generated_features, style_features):
    style_loss = 0.0
    for gen_feat, style_feat in zip(generated_features, style_features):
        # G_gen = gram_matrix(gen_feat)
        # G_style = gram_matrix(style_feat)

        G_gen = gen_feat.view(gen_feat.shape[0],-1)
        G_style = style_feat.view(style_feat.shape[0],-1)
        wd,_,_=WD(G_gen,G_style)
        # wd,_,_=WD(gen_feat,style_feat)

        style_loss += wd
    return style_loss/5



state_dict=torch.load('backbone/supcon.pth')['model']

for k in list(state_dict.keys()):
    # retain only encoder_q up to before the embedding layer
    if k.startswith("module.encoder"):
        state_dict[k[len("module.encoder.") :]] = state_dict[k]
    # delete renamed or unused k
    del state_dict[k]

model = models.__dict__['resnet50']()
model.load_state_dict(state_dict,strict=False)

# model = models.resnet50(pretrained=True)
feature_extractor = ResNetFeatureExtractor(model)
feature_extractor.eval()


generated_image=torch.rand(8,3,16,16)

style_image=torch.rand(8,3,16,16)

with torch.no_grad():
    # 假设这些是从生成图像和风格图像提取的特征
    generated_features = feature_extractor(generated_image)

    style_features = feature_extractor(style_image)

    # 计算风格损失
    style_loss = compute_style_loss(generated_features, style_features)
```

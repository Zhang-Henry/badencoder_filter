```python
import numpy as np
import matplotlib.pyplot as plt

# 读取文件
with open('results/cifar10/attack/cifar10_attack_output_trojan.txt', 'r') as file:
    entropy_trojan_trojan = file.read()
with open('results/cifar10/attack/cifar10_attack_output_clean.txt', 'r') as file:
    entropy_trojan_clean = file.read()
```

```python
entropy_trojan_trojan=[float(i) for i in entropy_trojan_trojan.split(' ')]
entropy_trojan_clean=[float(i) for i in entropy_trojan_clean.split(' ')]
```

```python
bins = 30

# 画直方图
plt.hist(entropy_trojan_trojan, bins, weights=np.ones(len(entropy_trojan_trojan)) / len(entropy_trojan_trojan), alpha=0.5, label='with trojan')
plt.hist(entropy_trojan_clean, bins, weights=np.ones(len(entropy_trojan_clean)) / len(entropy_trojan_clean), alpha=0.5, label='without trojan')

plt.legend(loc='upper right', fontsize = 20)
plt.ylabel('Probability (%)', fontsize = 20)
plt.title('normalized entropy', fontsize = 20)
plt.tick_params(labelsize=20)

fig1 = plt.gcf()
plt.show()
```

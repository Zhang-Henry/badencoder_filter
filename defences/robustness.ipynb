```python
import matplotlib.pyplot as plt
import numpy as np

# Data
ablate_ba = [38.67, 38.61, 38.63, 45.26, 53.31, 61.3]
ablate_asr = [87, 65.12, 34.93, 99.15, 100, 100]
filter_ba = [38.47, 38.03, 38.46, 45.52, 53.2, 60.95]
filter_asr = [99.85, 99.59, 100, 99.92, 100, 100]

# X-axis labels (assuming these are consistent for both methods)
x_labels = ['Label1', 'Label2', 'Label3', 'Label4', 'Label5', 'Label6']

# Plotting
plt.figure(figsize=(10, 6))

# Ablate - BA and ASR
plt.plot(x_labels, ablate_ba, marker='o', label='Ablate - BA', color='blue')
plt.plot(x_labels, ablate_asr, marker='o', label='Ablate - ASR', color='red')

# Filter - BA and ASR
plt.plot(x_labels, filter_ba, marker='x', linestyle='--', label='FilterAttack - BA', color='blue')
plt.plot(x_labels, filter_asr, marker='x', linestyle='--', label='FilterAttack - ASR', color='red')

# Labeling
plt.xlabel('X-axis Label')
plt.ylabel('Values')
plt.title('Comparison of Ablate and Filter Methods')
plt.legend()

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
def plot(ablate_ba, ablate_asr, filter_ba, filter_asr, x_labels, noise, xlabel):
    # Plotting
    plt.figure(figsize=(5, 5))

    # Ablate - BA and ASR
    plt.plot(x_labels, ablate_ba, marker='o', label='Ablate color - BA', color='blue')
    plt.plot(x_labels, ablate_asr, marker='o', label='Ablate color - ASR', color='red')

    # Filter - BA and ASR
    plt.plot(x_labels, filter_ba, marker='x', linestyle='--', label='FilterAttack - BA', color='blue')
    plt.plot(x_labels, filter_asr, marker='x', linestyle='--', label='FilterAttack - ASR', color='red')


    # Labeling
    plt.xlabel(xlabel, fontsize=14, fontweight='bold')
    if noise == 'JPEG compression quality':
        plt.ylabel('ASR / BA (%)', fontsize=14, fontweight='bold')
    # plt.title('Comparison of Ablate and Filter Methods under {}% Noise'.format(noise))
    plt.legend()
    plt.tight_layout()
    plt.savefig('robust/{}.png'.format(noise))
    # Show plot
    plt.show()
```

```python
# Data
ablate_ba = [61.71,59.07,56.42,55.77,53.31,48.83]
ablate_asr = [56.73,60.68,87.12,22.13,28.71,12.78]
filter_ba = [63.25,60.23,57.97,57.48,53.46,50.05]
filter_asr = [99.99,99.91,99.92,100,100,99.91]

# X-axis labels (assuming these are consistent for both methods)
x_labels = [2, 3, 4, 5, 7, 10]
plot(ablate_ba, ablate_asr, filter_ba, filter_asr, x_labels, 'Poisson Noise', 'Poisson Noise Intensity')
```

```python
import matplotlib.pyplot as plt
import numpy as np
def plot(ablate_ba, ablate_asr, filter_ba, filter_asr, x_labels, noise, xlabel):
    # Plotting
    plt.figure(figsize=(5, 5))

    # Ablate - BA and ASR
    plt.plot(x_labels, ablate_ba, marker='o', label='W/O disentangle loss - BA', color='#0c84c6')
    plt.plot(x_labels, ablate_asr, marker='o', label='W/O disentangle loss - ASR', color='#FA7F6F')

    # Filter - BA and ASR
    plt.plot(x_labels, filter_ba, marker='x', linestyle='--', label='Ours - BA', color='#0c84c6')
    plt.plot(x_labels, filter_asr, marker='x', linestyle='--', label='Ours - ASR', color='#FA7F6F')


    # Labeling
    plt.xlabel(xlabel, fontsize=14, fontweight='bold')
    if noise == 'JPEG compression quality':
        plt.ylabel('ASR / BA (%)', fontsize=14, fontweight='bold')
    # plt.title('Comparison of Ablate and Filter Methods under {}% Noise'.format(noise))
    plt.legend()
    plt.tight_layout()
    plt.savefig('robust/{}.png'.format(noise))
    # Show plot
    plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Function to plot a single subplot
def plot_subplot(ax, ablate_ba, ablate_asr, filter_ba, filter_asr, x_labels, title, xlabel):
    # Ablate - BA and ASR
    ax.plot(x_labels, ablate_ba, marker='o', label='W/O disentangle loss - BA', color='#0c84c6')
    ax.plot(x_labels, ablate_asr, marker='o', label='W/O disentangle loss - ASR', color='#FA7F6F')

    # Filter - BA and ASR
    ax.plot(x_labels, filter_ba, marker='x', linestyle='--', label='Ours - BA', color='#0c84c6')
    ax.plot(x_labels, filter_asr, marker='x', linestyle='--', label='Ours - ASR', color='#FA7F6F')

    # Labeling
    ax.set_xlabel(xlabel, fontsize=14)

    # ax.set_title(title)
    if xlabel=='JPEG Compression Quality':
        ax.set_ylabel('ASR / BA (%)', fontsize=14)
        ax.legend()


# Data for plots
data = [
    ([38.61, 45.26, 53.31, 61.3], [65.12, 99.15, 100, 100], [38.03, 45.52, 53.2, 60.95], [99.59, 99.92, 100, 100], [2,5,10,20], 'JPEG Compression Quality', 'JPEG Compression Quality'),
    ([61.71,59.07,55.77,53.31,48.83], [56.73,60.68,22.13,28.71,12.78], [63.25,60.23,57.48,53.46,50.05], [99.99,99.91,100,100,99.91], [2, 3, 5, 7, 10], 'Poisson Noise', 'Poisson Noise Intensity'),
    ([53.93,40.95,34.65,30.7], [99.75,74.3,36.54,13.22], [55.42,42.31,35.425,31.98], [98.57,92.93,87.83,79.01], [0.01,0.03,0.05,0.07], 'Salt and Pepper Noise', 'Salt and Pepper Noise Density')
]

# Creating a figure to hold the subplots
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Looping through data to create each subplot
for ax, (ablate_ba, ablate_asr, filter_ba, filter_asr, x_labels, title, xlabel) in zip(axs, data):
    plot_subplot(ax, ablate_ba, ablate_asr, filter_ba, filter_asr, x_labels, title, xlabel)

plt.tight_layout()
plt.savefig('robust/sensitive.pdf')
plt.show()
```
